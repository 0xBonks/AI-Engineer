# Virtual Environments

## Why Virtual Environments?

- Isolate dependencies per project
- Prevent version conflicts
- Reproducible environments

## Create and Use

```bash
# Create
python -m venv venv

# Activate (Linux/Mac)
source venv/bin/activate

# Activate (Windows)
venv\Scripts\activate

# Install packages
pip install openai

# Deactivate
deactivate
```

## Best Practices

1. One venv per project
2. Add venv/ to .gitignore
3. Use requirements.txt
4. Pin versions for reproducibility
